MATPLOTLIB

```python
plt.scatter(data.total_bill, data.tip)
```

```python
plt.scatter(data[data.smoker == 'Yes'].total_bill, data[data.smoker == 'Yes'].tip)
plt.scatter(data[data.smoker == 'No'].total_bill, data[data.smoker == 'No'].tip)
plt.xlabel('total_bill')
plt.ylabel('tip')
```

```python
plt.hist(data.total_bill, bins=100);
```

```python
plt.boxplot(data[data.day=='Sun'].total_bill);
```

```python
plt.imshow(data.corr())
plt.colorbar()
```

```python
from scipy.stats.distributions import alpha

alpha = 0.5
plt.scatter(data[data.smoker == 'Yes'].total_bill, data[data.smoker == 'Yes'].tip, alpha=alpha, label='Smoker')
plt.scatter(data[data.smoker == 'No'].total_bill, data[data.smoker == 'No'].tip, alpha=alpha, label='No Smoker')

plt.title('smoker')
plt.xlabel('total_bill')
plt.ylabel('tip')

plt.grid()
plt.legend()
```
